```python
import pandas as pd
#loading the dataset
dataset = pd.read_csv("preplacement.csv")
def stdNBgraph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    # list of column names is stored as 'values'
    values=[i for i in dataset]
    # z-scores are calculated for each value using (value - mean) / std
    z_score=[(j-mean)/std for j in values]
  # histogram with KDE curve is plotted
    sns.distplot(z_score,kde=True)
    # mean of the z-scores is calculated (should be ~0)
    sum(z_score)/len(z_score)
    print("Mean of z-scores:", sum(z_score)/len(z_score))
```

```python
stdNBgraph(dataset["ssc_p"])
```
